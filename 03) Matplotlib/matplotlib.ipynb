```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

## two ways to create matplotlib plot i) functional method ii) object oriented method, the ii) is better

```python
# Functional Method
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('This is Title')
```

```python
plt.subplot(1,2,1)  # multiplots, arguments (no of rows, column, plot no you are referring to)
plt.plot(x,y,'r')   # 'r' is red, 'b' is blue

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Method
fig = plt.figure()                     # figure object has been created

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # creates blank box, takes list as argument [left of the axes, bottom of the axes, width, height of the axes]

axes.plot(x,y)                         # quirk of jupyter notebook all should be in same block(says jose portila)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('The syntax is slightly different from above')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # range bw 0 & 1
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # 20% from the left, 50% from the bottom, 40% is the width and 30% is the height
```

```python
fig = plt.figure()                        # created empty figure

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])   # added a set of axis 
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])   # added another set of axis

axes1.plot(x,y)                           # ploting the axis
axes1.set_title('Outer large plot')      

axes2.plot(y,x)
axes2.set_title('Inner small plot')
```

```python
# creating subplots using OO method
fig,axes = plt.subplots()

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:   # iterating through axes array
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)                # using index calls
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()               # will take care of any overlapping plot(recommend to put at the end of the plot statements)
```

## Figure size, aspect ratio, DPI(dots per inch or pixels per inch)

```python
fig = plt.figure(figsize=(8,2))  # tuple which has width and height of the figure in inches(8 by 2 inches), to change/input dpi: fig = plt.figure(figsize=(3,2), dpi=100)  

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(figsize=(8,2))
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(x,y)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3)
```

```python
# adding a legend: to clarify which plot is which,,, an extra label argument is passed
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Squared')
ax.plot(x,x**3, label = 'X Cubed')

ax.legend() # ax.legend(loc=0) , choses the best location for the postion of legend, here It is already in best position(not overlapping the curve) so I didnt passed loc=0. Can also pass a tuple ie loc = (0.1,0.1)
```

## Plot appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='magenta', alpha=0.5)  # ax.plot(x,y, color='#FF8C00')  ie can use hex code as well,, alpha is transparency
```

```python
ax.plot(x,y, color='magenta', linewidth=5) # default thickness is 1 ,, lw can be typed instead of linewidth
fig
```

```python
# we need to write again and again to see variations(tweak fo jupyter)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='magenta', lw = 3, linestyle='--')  # linestyle =':' , linestyle ='-.' etc can use ls instead of linestyle
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='magenta', lw = 3, ls='-', marker='o' ) # mark the X points
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='magenta', lw = 3, ls='-', marker='*', markersize=20, 
        markerfacecolor='yellow', markeredgewidth = 3, markeredgecolor='grey')   # refer to the notebook for more
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='magenta', lw = 3, ls='--')

ax.set_xlim([0,1])  # I want to show plot bw 0 and 1 in the x axis
ax.set_ylim([0,2])  # I want to show plot bw 0 and 2 in the y axis
```
