```python
# Unsupervised Learning
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
data
```

```python
data[0]
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0],data[0][:,1]) # grabing all the rows and first column, all the rows and second column
```

```python
data[1]   # correct labels
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5)   # we can change this cluster number and run again 4,3,2,,,8,9 any no of clusters
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_   # predicted labels
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))  # tuple unpacking to grab the axis

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
