```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('df1', index_col=0)
df1.head()    
# the index is actually a time series. (2000-01-01 and so on, this index is a time series)
```

```python
df2 = pd.read_csv('df2')
df2.head()   
# index is non - time series
```

```python
df1['A'].hist()  # you want histogram of all the values in A column of df1
```

```python
df1['A'].hist(bins = 30) # we can pass such arguments cause it calls matplotlib under the hood
```

```python
df1['A'].plot(kind='hist', bins = 50)  # alternate method
# histograms are used the most in pandas
```

```python
df1['A'].plot.hist()  # another method
```

```python
df2.plot.area()
```

```python
df2.plot.area(alpha = 0.4)
```

```python
df1.plot.hist()  # we will use this method more
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked = True)   # calling pandas for quick analysis else use seaborn
```

```python
df1.plot.line(y='B', figsize=(12,3))  # the latest version of pandas automatically takes the index in x axis
```

```python
df1.plot.scatter(x='A', y='B', c='C') # third argument c is color displayed with the help of column C: 3D type plot
```

```python
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

```python
df1.plot.scatter(x='A', y='B', s=df1['C']*100) # s is size
```
