## Distribution Plots

```python
tips = sns.load_dataset('tips')  # seaborn comes with some inbuilt dataets
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])   # shows the distribution of 1 variable
```

```python
sns.distplot(tips['total_bill'], kde = False)  # kde is kernel density estimation
```

```python
sns.jointplot(x='total_bill',y='tip' ,data=tips)  # when we want to compare two variables.
```

```python
sns.jointplot(x='total_bill',y='tip' ,data=tips, kind = 'hex') # by default it is scattered, hex makes hexagons
```

```python
sns.jointplot(x='total_bill',y='tip' ,data=tips, kind = 'reg')  # reg is regression
```

```python
sns.jointplot(x='total_bill',y='tip' ,data=tips, kind = 'kde')  # shows the density where the points matches the most
```

```python
sns.pairplot(tips) # just pass the data name and it will perform jointplot for every possible combination
```

```python
sns.pairplot(tips, hue='sex')  # pass a categorical column here sex has two categories male and female
```

```python
sns.rugplot(tips['total_bill'])  # dash mark for every single point on the distribution line
```

## Categorial Plots

```python
import seaborn as sns 
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()   # head prints first 5 data points.
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)  # one of variable out of x and y should be numeric, by default it shows the average or mean
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)  # we want to compare the standard deviation of both, can pass your own aggregate column
```

```python
sns.countplot(x='sex', data=tips)  #explicitely counts the no of occurrences(which is the already chosen y axis)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

### box plot is better for normal audience, violin plot can be made for advance audience having knowledge of data science

```python
sns.stripplot(x='day', y='total_bill', data= tips) # one variable is categorial(x), y is ; default value of jitter is True
```

```python
sns.stripplot(x='day', y='total_bill', data= tips, jitter=False)  # this is of less use, the default is better
```

```python
sns.stripplot(x='day', y='total_bill', data= tips, hue='sex', split=True)
```

```python
# violin + strip = swarm plot
sns.swarmplot(x='day', y='total_bill', data= tips)  # points are adjusted so that they don't overlap; 1 drawback is their unability to scale well with large numbers
# as people are less used to see swarm plots better you can show violin and swarm together as below
```

```python
sns.violinplot(x='day', y='total_bill', data= tips)
sns.swarmplot(x='day', y='total_bill', data= tips, color='black')
```

```python
sns.catplot(x='day', y='total_bill', data= tips, kind='box')  # you can specify the kind here as box, violin, swarm etc
```

## Matrix Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
tc=tips.corr()
tc
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc, annot= True)
```

```python
sns.heatmap(tc, annot= True, cmap='coolwarm')
```

```python
flights
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers') # now we will have our data in some sort of matrix form
fp
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp, cmap='magma', linecolor='white', linewidth='2')
```

```python
sns.clustermap(fp)  # hierarchial clustering to produce a clustered version of the heatmap
# tries to cluster the information which are similar to each other
```

```python
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)  # we will understand mathematics behind these methods in ML part.
```

## Grids

```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')   # measurements of bunch of different flowers
iris.head()
```

```python
iris['species'].unique()  # can see that there are 3 different species.
```

```python
sns.pairplot(iris)  # automated joint plot for this dataset
```

```python
g = sns.PairGrid(iris)
g
```

```python
g = sns.PairGrid(iris)
import matplotlib.pyplot as plt
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.histplot, 'total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')   # this takes two arguments so we pass it
```

## Regression Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')   # measurements of bunch of different flowers
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)   # L-M plot to make simple linear model plot
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'], scatter_kws={'s':75}) # changing the size, just passing any dictionary with base matplotlib arguments s for size of a scatterplot
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')  # instead of separating by color we separated them by column
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```
